```python
# =========================================================
# Problem 1: Post Engagement Analyzer
# =========================================================

def analyze_post(likes_list):

    total_likes = 0   # variable to store total likes

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check viral condition
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example data
likes = [200, 300, 250, 300]
analyze_post(likes)
```

```python
# =========================================================
# Problem 2: Medicine Stock Alert
# =========================================================

def check_medicine_stock(stock):

    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


check_medicine_stock(6)
```

```python
# =========================================================
# Problem 3: Rainfall Adequacy Checker
# =========================================================

def check_rainfall(rainfall_data, required_level):

    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rain_data = [60, 70, 80, 78]
check_rainfall(rain_data, 70)
```

```python
# =========================================================
# Problem 4: Duplicate Account Detection
# =========================================================

def check_duplicates(usernames):

    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["user1", "user2", "user1"]
check_duplicates(users)
```

```python
# =========================================================
# Problem 5: Appointment Eligibility Checker
# =========================================================

def check_eligibility(age):

    print("Patient Age:", age)

    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")


check_eligibility(21)
```

```python
# =========================================================
# Problem 6: Premium Crop Price Filter
# =========================================================

def filter_premium_crops(prices):

    premium = []   # list to store premium prices

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# =========================================================
# Problem 7: Application Health Checker
# =========================================================

def check_system_health(error_count):

    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


check_system_health(7)
```

```python
# =========================================================
# Problem 8: Transaction Limit Checker
# =========================================================

def check_transaction(amount):

    daily_limit = 50000

    print("Transaction Amount:", amount)

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


check_transaction(60000)
```

```python
# =========================================================
# Problem 9: Attendance Eligibility System
# =========================================================

def check_exam_eligibility(attendance_list):

    present = 0

    for day in attendance_list:
        if day == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

```python
# =========================================================
# Problem 10: Smart Electricity Bill Analyzer
# =========================================================

def calculate_electricity_bill(units):

    bill = 0   # variable to store total bill

    # Calculate bill using slab system
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify usage
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)


calculate_electricity_bill(250)
```
