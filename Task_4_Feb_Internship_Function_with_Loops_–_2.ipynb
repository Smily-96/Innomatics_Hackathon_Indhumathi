```python
# =========================================================
# Mobile Recharge Validation System
# =========================================================

# Function to validate recharge
def validate_recharge():

    valid_plans = [199, 299, 399, 599]   # list of valid recharge plans

    while True:   # this loop allows user to retry if invalid

        amount = int(input("Enter recharge amount: ₹"))   # take recharge input

        # Check minimum recharge rule
        if amount < 50:
            print("Invalid! Minimum recharge must be ₹50.")

        # Check if plan exists in valid plans list
        elif amount not in valid_plans:
            print("Invalid Plan! Available plans are:", valid_plans)

        # If all conditions are satisfied
        else:
            print("Recharge Successful for ₹", amount)
            break   # exit loop after successful recharge


# Call the function
validate_recharge()
```

```python
# =========================================================
# Inventory Reorder Alert System
# =========================================================

# Function to check stock levels
def check_inventory(products):

    # Loop through each product and its stock quantity
    for product, stock in products.items():

        # Check if stock is less than 15
        if stock < 15:
            print(product, ":", "Reorder Alert")   # Low stock

        else:
            print(product, ":", "Stock OK")        # Sufficient stock


# Example dictionary of products and their stock
inventory = {
    "Shampoo": 20,
    "Soap": 10,
    "Toothpaste": 8,
    "Oil": 25
}

# Call the function
check_inventory(inventory)
```

```python
# =========================================================
# Student Result Processing System
# =========================================================

# Function to process student marks
def process_result(marks):

    total = 0   # variable to store total marks

    # Loop through each mark in the list
    for mark in marks:
        total = total + mark   # add each mark to total

    # Calculate average
    average = total / len(marks)

    # Check pass or fail condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example list of student marks
student_marks = [60, 70, 55, 40]

# Call the function and store result
result = process_result(student_marks)

# Print the result
print("Student Result:", result)
```

```python
# =========================================================
# Cab Fare Estimator with Retry Option
# =========================================================

# Function to calculate cab fare
def calculate_fare():

    base_fare = 50        # fixed base fare
    rate_per_km = 12      # cost per kilometer

    while True:   # loop to allow retry

        # Take distance input from user
        distance = float(input("Enter distance in km: "))

        # Ask if it is peak hour
        peak_hour = input("Is it peak hour? (yes/no): ")

        # Calculate normal fare
        fare = base_fare + (distance * rate_per_km)

        # Add 25% extra if peak hour
        if peak_hour.lower() == "yes":
            fare = fare * 1.25   # increase fare by 25%

        # Print total fare
        print("Total Fare: ₹", round(fare, 2))

        # Ask user if they want to retry
        retry = input("Do you want to calculate again? (yes/no): ")

        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break   # exit loop


# Call the function
calculate_fare()
```

```python
# =========================================================
# Employee Attendance Eligibility Checker
# =========================================================

# Function to check attendance eligibility
def check_attendance(attendance_list):

    present_count = 0   # variable to count number of present days

    # Loop through each day in attendance list
    for day in attendance_list:

        # Check if the day is Present
        if day == "P":
            present_count = present_count + 1   # increase present count

    # Calculate attendance percentage
    percentage = (present_count / len(attendance_list)) * 100

    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance data
attendance_data = ["P", "P", "A", "P", "P"]

# Call the function and store result
result = check_attendance(attendance_data)

# Print final result
print("Attendance Eligibility:", result)
```

```python
# =========================================================
# Password Strength Checker
# =========================================================

# Function to check password strength
def check_password(password):

    has_digit = False        # flag to check if password has a digit
    has_special = False      # flag to check if password has special character
    special_characters = "@#$"   # allowed special characters

    # Rule 1: Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # Loop through each character in password
    for char in password:

        # Check if character is a digit
        if char.isdigit():
            has_digit = True

        # Check if character is one of the special characters
        if char in special_characters:
            has_special = True

    # Final condition check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include at least one digit and one special character)"


# Example test passwords
print(check_password("Pass123"))      # weak (no special character)
print(check_password("Password@1"))  # strong
```
