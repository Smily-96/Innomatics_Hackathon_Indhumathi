1. Problem Statement:

Given a username and password, check whether login is successful.

username = "admin"

password = "1234"

Requirements:

Print "Login Successful" if both username and password match

Otherwise print "Invalid Credentials"

Real-World Application: Authentication systems


```python
# ==========================================
# TASK 1 – User Login Check
# ==========================================

# Correct login details stored in system
correct_username = "admin"
correct_password = "1234"


# -------- First Login Attempt (Correct) --------
print("First Login Attempt:")

username1 = "admin"
password1 = "1234"

if username1 == correct_username and password1 == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")


# -------- Second Login Attempt (Wrong Password) --------
print("\nSecond Login Attempt:")

username2 = "admin"
password2 = "1111"   # wrong password

if username2 == correct_username and password2 == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

Problem Statement:

Analyze student results.

marks = [45, 78, 90, 33, 60]

Requirements:

A student passes if marks ≥ 50

Count the total number of pass students

Count the total number of fail students

Print the final result clearly

Real-World Application: Academic evaluation systems


```python
# ==========================================
# TASK 2 – Pass / Fail Analyzer
# ==========================================

# List of student marks
marks = [45, 78, 90, 33, 60]

# Counters for pass and fail students
pass_count = 0
fail_count = 0

# Loop through each mark
for mark in marks:
    if mark >= 50:        # Condition for pass
        pass_count = pass_count + 1
    else:                 # Otherwise fail
        fail_count = fail_count + 1

# Print final result
print("\nTotal Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

3. Simple Data Cleaner

Problem Statement:

Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]

Requirements:

Remove extra spaces

Convert all names to lowercase

Print the cleaned list

Real-World Application: Data preprocessing before analysis


```python
# ==========================================
# TASK 3 – Simple Data Cleaner
# ==========================================

# List of names with spaces and different cases
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Loop through each name
for name in names:
    name = name.strip()      # Remove extra spaces
    name = name.lower()      # Convert to lowercase
    cleaned_names.append(name)  # Add cleaned name to list

# Print cleaned list
print("\nCleaned Names:", cleaned_names)
```

4. Message Length Analyzer

Problem Statement:

Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:

Print the length of each message

Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems


```python
# ==========================================
# TASK 4 – Message Length Analyzer
# ==========================================

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Check each message
for message in messages:
    length = len(message)   # Find length of message
    print("\nMessage:", message)
    print("Length:", length)

    # Flag if message length is greater than 10
    if length > 10:
        print("This message is longer than 10 characters")
```

5. Error Message Detector

Problem Statement:

Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:

Count the number of "ERROR" entries

Print the total error count

Real-World Application: Monitoring and log analysis systems


```python
# ==========================================
# TASK 5 – Error Message Detector
# ==========================================

# System log list
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Variable to count errors
error_count = 0

# Loop through logs
for log in logs:
    if log == "ERROR":
        error_count = error_count + 1

# Print total error count
print("\nTotal ERROR messages:", error_count)
```
