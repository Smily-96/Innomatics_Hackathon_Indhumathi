Problem Statement 1: Employee Performance Bonus Eligibility

```python
# =========================================================
# Problem 1: Employee Performance Bonus Eligibility
# =========================================================

# Dictionary containing employee names and performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
highest_score = max(employees.values())

# Find all employees who have this highest score (handle ties)
top_performers = []

for name, score in employees.items():   # loop through dictionary
    if score == highest_score:          # check if score equals highest
        top_performers.append(name)     # add employee to list

# Print result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")

```

Problem Statement 2: Search Query Keyword Analysis

```python
# =========================================================
# Problem 2: Search Query Keyword Analysis
# =========================================================

# Given search query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Split sentence into words
words = query.split()

# Dictionary to store word frequency
frequency = {}

# Count frequency of each word
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Display only words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print("Repeated keywords:", result)

```

Problem Statement 3: Sensor Data Validation

```python
# =========================================================
# Problem 3: Sensor Data Validation
# =========================================================

# Given sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings as (index, value)
valid_readings = []

# Loop through readings with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:                 # check if number is even
        valid_readings.append((index, value))   # store pair

# Print valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)


```

Problem Statement 4: Email Domain Usage Analysis

```python
# =========================================================
# Problem 4: Email Domain Usage Analysis
# =========================================================

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to count domain usage
domain_count = {}

# Extract domain and count
for email in emails:
    domain = email.split("@")[1]   # get text after '@'

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Total number of emails
total_emails = len(emails)

# Print percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(domain + ":", int(percentage), "%")

```

Problem Statement 5: Sales Spike Detection

```python
# =========================================================
# Problem 5: Sales Spike Detection
# =========================================================

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Print days where sales are more than 30% above average
for index, value in enumerate(sales):
    if value > average_sales * 1.3:   # 30% above average
        print("Day", index + 1, ":", value)
```

Problem Statement 6: Duplicate User ID Detection

```python
# =========================================================
# Problem 6: Duplicate User ID Detection
# =========================================================

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to count occurrences
user_count = {}

for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Print only duplicates
for user, count in user_count.items():
    if count > 1:
        print(user, "→", count, "times")
```
